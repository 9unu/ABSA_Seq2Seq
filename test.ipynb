```python
from kss import split_sentences

text = '갤럭시같이 맥세이프 호환이 안되는 기종과도 보조배터리를 공유할 수 있어서 좋을 것 같네요 5000으로 구매했는데 무게는 그럭저럭 가볍습니다 솔직하게 발열은 좀 있는 편이에요 택배 도착 후 뜯어서 바로 충전해보는데 10분도 안지나서 아래쪽에서 발열감이 느껴져요 엄청 뜨거운 건 아니고 휴대폰 오래 사용 시 느낄 수 있는 발열감 정도 되는 것 같네요 충전은 1분에 15퍼정도 되는 것 같습니다 정확하게 측정한건 아니라서 참고만 해주세요'
ignores = [
    '갤럭시같이 맥세이프 호환이 안되는 기종과도 보조배터리를 공유할 수 있어서 좋을 것 같네요',
    '5000으로 구매했는데 무게는 그럭저럭 가볍습니다',
    '솔직하게 발열은 좀 있는 편이에요 택배 도착 후 뜯어서 바로 충전해보는데 10분도 안지나서 아래쪽에서 발열감이 느껴져요 엄청 뜨거운 건 아니고 휴대폰 오래 사용 시 느낄 수 있는 발열감 정도 되는 것 같네요',
    '충전은 1분에 15퍼정도 되는 것 같습니다',
    '일반 고속 충전 케이블로 충전하는 정도보다 약간 느리게 나오는 듯 합니다',
    '접착력은 엄청 강합니다 한손으로 잡고 세게 흔들어도 안떨어져요'
]

def preprocess_text(text, ignores):
    for ignore in ignores:
        text = text.replace(ignore, f"|||{ignore}|||")
    return text

def postprocess_sentences(sentences):
    processed_sentences = []
    for sentence in sentences:
        sentence = sentence.replace("|||", "")
        processed_sentences.append(sentence)
    return processed_sentences

def custom_split_sentences(text, ignores):
    preprocessed_text = preprocess_text(text, ignores)
    sentences = split_sentences(preprocessed_text, backend="auto")
    filtered_sentences = postprocess_sentences(sentences)
    return filtered_sentences

result = custom_split_sentences(text, ignores)
print(result)
```

```python
import os
import json
folder_path = "json_data"

def load_json_files(folder_path):
    file_list = os.listdir(folder_path)
    all_reviews = []
    for file_name in file_list:
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)
            with open(file_path, 'r', encoding='utf-8-sig') as file:
                data = json.load(file)
                all_reviews.extend(data)
    return all_reviews
```

```python
df=load_json_files(folder_path)
```

```python
import pandas as pd
df = pd.json_normalize(df)
```

```python
df['content'] = df['content'].apply(lambda row: row.replace("\n", ""))
```
